```python
rows = []
with open('filtered_stars_with_gravity.csv', 'r') as f:
  csv_reader = csv.reader(f)
  for row in csv_reader:
    rows.append(row)

headers = rows[0]
star_data_rows = rows[1:]
```

```python
final_star_dict = []

for star_data in star_data_rows:
  temp_dict = {
                  "star_name": star_data[1],
                  "distance_from_earth": star_data[2],
                  "star_mass": star_data[3],
                  "star_radius": star_data[4],
                  "gravity": star_data[5],
              }
  final_star_dict.append(temp_dict)

print(final_star_dict)
```
